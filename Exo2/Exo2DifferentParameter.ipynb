```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Generate data
x = np.linspace(-1, 1, 100)
beta = [.1, .1, 1]
tnoisy = beta[0] + beta[1]*x + beta[2]*x**2 + np.random.normal(0, .1, len(x))

# Reshape x to be a 2D array
x = x.reshape(-1, 1)

# 1. Generate polynomial features
degree = 30  # We are using quadratic polynomial (degree 2)
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(x)

# 2. Fit a linear model to the polynomial features
model = LinearRegression()
model.fit(X_poly, tnoisy)

# 3. Predict on the test set
xtest = np.linspace(-1, 1, 100).reshape(-1, 1)
X_polytest = poly_features.transform(xtest)
y_pred = model.predict(X_polytest)

# Plot the original noisy data
plt.scatter(x, tnoisy, c='r', label='Noisy Data')

# Plot the predictions superimposed on the data
plt.plot(xtest, y_pred, c='b', label='Model Prediction')

plt.legend()
plt.title('Polynomial Regression Fit')
plt.show()
```
